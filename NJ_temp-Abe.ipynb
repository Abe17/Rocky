```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13)
x_axis_data
```

```python
# Average weather temp
ave_temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Plot the line
plt.plot(x_axis_data, ave_temp)
plt.show()
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_c = [(x-32) * 0.56 for x in ave_temp]
points_c
```

```python
# Plot using Celsius
plt.plot(x_axis_data, points_c)
plt.plot(x_axis_data, ave_temp)
plt.show()
```
